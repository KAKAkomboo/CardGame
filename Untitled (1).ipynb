```python
import pandas as pd
import seaborn as sb
import wordcloud as wc
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SMSSpamCollection", sep="\t", names = ["Label", "Text"])
df
```

```python
df["Label"].value_counts()
```

```python
dataset = df.to_dict("records")
index = int(len(dataset)*0.8)
train = dataset[:index]
text = dataset[index:]
```

```python
spam_words = {}
ham_words = {}

spam_text = ""
ham_text = ""

for item in train:
    label = item['Label']
    text = item['Text']
    text = text.lower()
    text = text.replace(",", "").replace(".", "").replace("!", "").replace("?", "")\
        .replace("-", "").replace(":", "").replace(";", "").replace("'", "").replace('"', "")

    if label == "spam":
        spam_text += "\n" + text
    else:
        ham_text += "\n" + text


    words = text.split(' ')

    for word in words:
        if label  == "spam":
            if word in spam_words:
                spam_words[word]+=1
            else:
                spam_words[word]=1
        else:
            if word in ham_words:
                ham_words[word] += 1
            else:
                ham_words[word] = 1
```

```python
spam_wc = wc.WordCloud()
spam_wc.generate(spam_text)
plt.imshow(spam_wc)
```

```python
ham_wc = wc.WordCloud()
ham_wc.generate(ham_text)
plt.imshow(ham_wc)
```
